- # Advanced selection

  Now let's use boolean operators to create more advanced expressions
6. How many Fire-Flying Pokemons are there?


```python
# Try your code here

Fire_Flying_Pokemon = df[(df["Type 1"] == "Fire") & (df["Type 2"] == "Flying")]
Fire_Flying_Pokemon["Name"].count()
```

7. How many 'Poison' pokemons are across both types?

```python
# Try your code here

Poison_Pokemon = df[(df["Type 1"] == "Poison") | (df["Type 2"] == "Poison")]
Poison_Pokemon["Name"].count()
```

8. Name the pokemon of Type 1 Ice which has the strongest defense?

```python
# Try your code here

Stongest_Ice_Pokemon = df[df["Type 1"] == "Ice"].sort_values(by="Defense", ascending=False)
print(Stongest_Ice_Pokemon.iloc[0])
```

9. What's the most common type of Legendary Pokemons?

```python
# Try your code here

Common_Legendary_Pokemon_by_Type = df[(df["Legendary"] == True) & (df["Type 1"])].mode()
Common_Legendary_Pokemon_by_Type
```

```python
# An easier solution from the website datawars 😅:
df.loc[df['Legendary'], 'Type 1'].value_counts()
```

10. What's the most powerful pokemon from the first 3 generations, of type water?

```python
# Try your code here

Strongest_Defense_3rd_Gen_Water_Pokemon = df[(df["Type 1"] == "Water") & (df["Generation"] == 3)].sort_values(by="Total", ascending=False)
Strongest_Defense_3rd_Gen_Water_Pokemon
```

11. What's the most powerful Dragon from the last two generations?

```python
# Try your code here

Strongest_Dragon_Last_2_Gen = df[((df["Type 1"] == "Dragon") | (df["Type 2"] == "Dragon")) & (df["Generation"] >= 5)].sort_values(by="Total", ascending=False)
Strongest_Dragon_Last_2_Gen
```

12. Select most powerful Fire-type pokemons

```python
# Try your code here

powerful_fire_df = df[(df["Type 1"] == "Fire") & (df["Attack"] > 100)]
powerful_fire_df
```

13. Select all Water-type, Flying-type pokemons


```python
# Try your code here

water_flying_df = df[(df["Type 1"] == "Water") & (df["Type 2"] == "Flying")]
water_flying_df
```

14. Select specific columns of Legendary pokemons of type Fire


```python
# Try your code here

legendary_fire_df = df.loc[(df["Type 1"] == "Fire") & (df["Legendary"] == True), ["Name","Attack","Generation"]]
legendary_fire_df
```

15. Select Slow and Fast pokemons

  This is the distribution of speed of the pokemons. The red lines indicate those bottom 5% and top 5% pokemons by speed:


```python
ax = df['Speed'].plot(kind='hist', figsize=(10, 5), bins=100)

ax.axvline(df['Speed'].quantile(.05), color='red')
ax.axvline(df['Speed'].quantile(.95), color='red')
```

```python
# Try your code here

slow_fast_df = df[(df["Speed"] < df["Speed"].quantile(.05)) | (df["Speed"] > df["Speed"].quantile(.95))]
slow_fast_df
```

16. Find the Ultra Powerful Legendary Pokemon
